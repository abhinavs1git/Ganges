# Exploratory Spatial Data Analysis (ESDA)

This notebook explores the raw metrics for the Ganga Basin: Elevation, Rainfall, and Rivers.

```python
import geopandas as gpd
import matplotlib.pyplot as plt
import os

base_dir = '../data/raw'
# Load data
rivers = gpd.read_file(os.path.join(base_dir, 'rivers.geojson'))
grid = gpd.read_file(os.path.join(base_dir, 'basin_grid.geojson'))

print('Rivers:', rivers.shape)
print('Grid points:', grid.shape)
```

## Visualizing Initial Layers

```python
fig, ax = plt.subplots(figsize=(10, 8))
grid.plot(column='elevation_meters', ax=ax, legend=True, cmap='terrain', alpha=0.5)
rivers.plot(ax=ax, color='blue', linewidth=2)
ax.set_title('Elevation Map with Major Rivers (Ganga Basin)')
plt.show()
```

## Rainfall Distribution

```python
fig, ax = plt.subplots(figsize=(10, 8))
scatter = grid.plot(column='rainfall_mm', ax=ax, legend=True, cmap='Blues', alpha=0.7, markersize=10)
rivers.plot(ax=ax, color='blue', linewidth=1)
ax.set_title('Rainfall Map (mm) across Uttar Pradesh')
plt.show()
```
