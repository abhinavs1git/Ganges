# Interactive Visualizations and Heatmaps

Generating final risk map outputs for decision makers.

```python
import geopandas as gpd
import folium

results = gpd.read_file('../outputs/flood_risk_results.geojson')

# Calculate center
bounds = results.total_bounds
y = (bounds[1] + bounds[3]) / 2
x = (bounds[0] + bounds[2]) / 2

m = folium.Map(location=[y, x], zoom_start=7)
results.explore(column='risk_level', cmap='RdYlGn_r', m=m)
m
```
