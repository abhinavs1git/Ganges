# Flood Risk Modeling

We apply a weighted spatial approach to determine flood likelihood.

```python
import geopandas as gpd
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler

gdf = gpd.read_file('../data/processed/analysis_ready.gpkg')
gdf.head()
```

## Standardizing Features
We use MinMaxScaler to cast all variables between 0-1.

```python
scaler = MinMaxScaler()
gdf['norm_rain'] = scaler.fit_transform(gdf[['rainfall_mm']])
gdf['norm_prox'] = 1 - scaler.fit_transform(gdf[['distance_to_river_km']])
gdf['norm_elev'] = 1 - scaler.fit_transform(gdf[['elevation_meters']])

gdf['risk_score'] = (0.4 * gdf['norm_rain']) + (0.3 * gdf['norm_prox']) + (0.3 * gdf['norm_elev'])
gdf['risk_score_100'] = gdf['risk_score'] * 100

gdf['risk_score_100'].describe()
```
